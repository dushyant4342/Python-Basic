```
#plotting using scatter plt
plt.scatter(x,y,c='b') #c for color

plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Graph in 2D")
plt.savefig("test.png")
```

```
# plt plot
y=x*x
plt.plot(x,y,'g*-', linewidth = 1.1, markersize = 12) #g-- give such curve
```

```
# Creating subplots 
plt.subplot(2,2,1)
plt.plot(x,y,'r*--')
plt.subplot(2,2,2)
plt.plot(x,y,'go')
plt.subplot(2,2,3)
plt.plot(x,y**3,'b--')
plt.subplot(2,2,4)
plt.plot(x,y**0.5,'y')
```

```
# Bar plot
x= [2,4,6]
y = [4,9,11]
plt.bar(x,y, align = 'center')

x2 = [3,7,9]
y2 = [14,8,10]
plt.bar(x2,y2,color = 'g')
```

```
#Boxplot
data = np.random.rand(5)
plt.boxplot(data)  # do remember about 25, 75 percentile
```

```
# Data to plot a pie chart
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.4, 0, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=False)

plt.axis('equal')
plt.show()
```
